# Introduction

data source: https://www.kaggle.com/datasets/paramvir705/airbnb-data

The data is about short-term rental listings from a platform like Airbnb. It contains detailed information about various properties available for rent, including their characteristics, host details, and review statistics

***This is Detail for each columns:***
- id: Unique identifier for the listing.
- log_price: Logarithm of the listing price.
- property_type: Type of property (e.g., Apartment, House).
- room_type: Type of room offered (e.g., Entire home/apt, Private room, Shared room).
- amenities: List of amenities provided in the property. Represented as a JSON-like string.
- accommodates: Number of guests the property can accommodate.
- bathrooms: Number of bathrooms in the property.
- bed_type: Type of bed provided (e.g., Real Bed, Futon).
- cancellation_policy: Type of cancellation policy (e.g., strict, moderate, flexible).
- cleaning_fee: Indicates whether a cleaning fee is charged (True or False).
- city: City where the property is located.
- description: Description of the property.
- first_review: Date of the first review received.
- host_has_profile_pic: Indicates whether the host has a profile picture (t for true).
- host_identity_verified: Indicates whether the host's identity is verified (t for true, f for false).
- host_response_rate: Host’s response rate to inquiries (e.g., 100%).
- host_since: Date since the host has been active on the platform.
- instant_bookable: Indicates whether the property is available for instant booking (t for true, f for false).
- last_review: Date of the most recent review.
- latitude: Latitude coordinate of the property’s location.
- longitude: Longitude coordinate of the property’s location.
- name: Name of the listing.
- neighbourhood: Neighborhood where the property is located.
- number_of_reviews: Total number of reviews received by the listing.
- review_scores_rating: Average rating score based on reviews.
- thumbnail_url: URL of the thumbnail image for the listing.
- zipcode: ZIP code of the property’s location.
- bedrooms: Number of bedrooms in the property.
- beds: Number of beds provided in the property.

````python
# เขียนฟังก์ชั่นสำหรับทำให้ข้อมูลใน zipcode อยู่ใน format เดียวกัน
import re
import math
zip_code_pattern = r'^\d{5}(-\d{4})?$'


def fix_zipcode(zipcode):
    if pd.isna(zipcode):  # ถ้าเป็น NaN
        return zipcode
    zipcode = str(zipcode).strip()  # แปลง Zipcode ให้เป็น string และลบช่องว่าง
    if '.' in zipcode:
        zipcode = re.sub(r'.0$', '', zipcode)  # ลบ .0 ที่ท้ายออก
    if re.fullmatch(r'\d{5}', zipcode):  # ตรวจสอบว่าเป็นตัวเลข 5 หลัก
        return zipcode
    elif re.fullmatch(r'\d{5}-\d{4}', zipcode):  # ตรวจสอบว่าเป็นรูปแบบ Zip+4
        return zipcode
    else:
        return None  # หรือสามารถกำหนดค่าอื่น ๆ ที่เหมาะสมหากรูปแบบไม่ถูกต้อง

airbnb_df['zipcode'] = airbnb_df['zipcode'].apply(fix_zipcode)
````

##Detect outlier

````python
numerical_columns = ['log_price', 'accommodates', 'bathrooms', 'number_of_reviews', 'review_scores_rating', 'bedrooms','beds']
````

````python
plt.figure(figsize=(20, 15))

for i, column in enumerate(numerical_columns, 1):
    plt.subplot(3, 3, i)
    sns.boxplot(data=airbnb_df[column])
    plt.title(f'Box Plot of {column}')
    plt.xlabel(column)

plt.tight_layout()
plt.show()
````

````python
outlier_columns = ['log_price', 'accommodates', 'bathrooms', 'bedrooms','beds']
````

````python
# filter outlier ของ column ที่กำหนด
import numpy as np

outliers = {}
for i in outlier_columns:
  Q1 = airbnb_df[i].quantile(0.25)
  Q3 = airbnb_df[i].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers[i] = airbnb_df[(airbnb_df[i] < lower_bound) | (airbnb_df[i] > upper_bound)]
````

````python
# plot การกระจายตัวของ log_price
plt.figure(figsize=(8, 6))
sns.histplot(data=airbnb_df, x='log_price', bins=30, kde=False)
plt.title('Histogram of log_price')
plt.xlabel('log_price')
plt.ylabel('Frequency')
plt.show()
````

````python
# หา lower, upper bound
Q1 = airbnb_df['log_price'].quantile(0.25)
Q3 = airbnb_df['log_price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
````

````python
# Capping outlier
airbnb_df['log_price'] = np.where(airbnb_df['log_price'] > upper_bound, upper_bound, airbnb_df['log_price'])
airbnb_df['log_price'] = np.where(airbnb_df['log_price'] < lower_bound, lower_bound, airbnb_df['log_price'])
````

## Test with Multiple Models

````python
from sklearn.linear_model import LinearRegression, Ridge, Lasso, ElasticNet
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import r2_score, mean_squared_error
from sklearn.pipeline import Pipeline

# List of regression models
models = {
    'Linear Regression': LinearRegression(),
    'Ridge Regression': Ridge(),
    'Lasso Regression': Lasso(),
    'ElasticNet': ElasticNet(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(),
    'Gradient Boosting': GradientBoostingRegressor(),
}

# Train and evaluate each model
results = {}
for model_name, model in models.items():
    # Train the model
    model.fit(X_train, y_train)
    # Make predictions
    y_pred = model.predict(X_test)
    # Evaluate the model
    mse = mean_squared_error(y_test, y_pred)
    results[model_name] = mse
    # print(f'{model_name} Mean Squared Error: {mse}')

    r_squared =  r2_score(y_test, y_pred)
    print(f'{model_name} model accuracy: {r_squared}')

# Display all results
print("\nModel Evaluation Results:")
for model_name, mse in results.items():
    print(f'{model_name}: {mse}')
````

**Lasso Regression and ElasticNet Negative R-squared**

For Lasso and ElasticNet, the R-squared values are extremely close to zero and negative. Here’s why:

Overregularization: Both Lasso and ElasticNet include regularization terms that can penalize the coefficients. If the regularization strength (lambda for Lasso, and a combination of lambda and alpha for ElasticNet) is too high, it can shrink the coefficients of the predictors to zero or near zero. This leads to a model that essentially predicts the mean of the target variable, resulting in an R-squared close to zero or negative.

Model Fit: When the R-squared value is negative, it means that the model is performing worse than a model that simply predicts the mean of the target variable. This can happen if the regularization is too strong, making the model unable to fit the data properly.

## Additional Testing with K-Fold Method`

````python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score, explained_variance_score
````

````python
#ทดสอบความถูหต้องของโมเดล แบ่งจำนวน 5 กอง เลือกมา1กอง มาเป็น Target 4 ตัวที่เหลือเป็นตัว Train
k = 5  # Number of folds
kf = KFold(n_splits=k, shuffle=True, random_state=42)
````

````python
mae_list = []
mse_list = []
rmse_list = []
r2_list = []
explained_variance_list = []
````

````python
for train_index, test_index in kf.split(X):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]

    # Train the model
    model = LinearRegression()
    model.fit(X_train, y_train)

    # Predict
    y_pred = model.predict(X_test)

    # Calculate metrics
    mae_list.append(mean_absolute_error(y_test, y_pred))
    mse_list.append(mean_squared_error(y_test, y_pred))
    rmse_list.append(mean_squared_error(y_test, y_pred, squared=False))
    r2_list.append(r2_score(y_test, y_pred))
    explained_variance_list.append(explained_variance_score(y_test, y_pred))
````

````python
mean_mae = np.mean(mae_list)
std_mae = np.std(mae_list)

mean_mse = np.mean(mse_list)
std_mse = np.std(mse_list)

mean_rmse = np.mean(rmse_list)
std_rmse = np.std(rmse_list)

mean_r2 = np.mean(r2_list)
std_r2 = np.std(r2_list)

mean_explained_variance = np.mean(explained_variance_list)
std_explained_variance = np.std(explained_variance_list)
````

````python
print(f"Mean Absolute Error (MAE): {mean_mae:.4f} ± {std_mae:.4f}")
print(f"Mean Squared Error (MSE): {mean_mse:.4f} ± {std_mse:.4f}")
print(f"Root Mean Squared Error (RMSE): {mean_rmse:.4f} ± {std_rmse:.4f}")
print(f"R-squared (R²): {mean_r2:.4f} ± {std_r2:.4f}")
print(f"Explained Variance Score: {mean_explained_variance:.4f} ± {std_explained_variance:.4f}")
````

From the result, we can see the evaluation values of K-fold method has been so close to the train_test_split method
